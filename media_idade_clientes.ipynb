```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
customers = pd.read_csv('data/dataset-customers.csv', sep = ';')
customers.drop_duplicates(subset ="id",keep = False, inplace = True)
```

```python
def distribuicao_freq(dataset, data, titulo, xlabel, ylabel):
    distribuicao = sns.displot(dataset, x = data)
    fig = distribuicao.fig 
    fig.suptitle(titulo, fontsize=12)
    distribuicao.set(xlabel= xlabel, ylabel= ylabel)

distribuicao_freq(customers, 'age', 'Distribuição da Frequência por Idade', 'Idade', 'Quantidade')
```

```python
print('A média da coluna age é igual {}, portanto a média de idade entre os cliente é aproximandamente 35 anos.'
      .format(round(customers['age'].mean(), 2)))
```

- **Conclusão:** A média de idades dos clientes é igual a 35.06, ou seja, aproximadamente 35 anos. Através do gráfico de frequência relativo a idades dos clientes é possível observar que apesar da média ter o valor de 35.06, a distribuição das idades é bem variada indicando que a média de idade não é uma variável significativa para identificar algum padrão no *dataset* de clientes, sendo inclusive o intervalo de 35 anos um dos intervalos que possue uma das menores quantidades de dados.
